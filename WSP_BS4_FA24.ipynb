```python
#BS4
#Website.
!pip install bs4
!pip install pandas
from bs4 import BeautifulSoup
from urllib.request import urlopen
import re
import pandas as pd
import pprint
```

```python
#Scraping website below
#https://scrapeme.live/shop/
html = urlopen('https://scrapeme.live/shop/')
bs_html = BeautifulSoup(html, 'html.parser')
print(bs_html)
```

```python
#Start with 1
cards = bs_html.find('ul', {'class':'products columns-4'})
```

```python
#For 1 case.
name = cards.find('h2', {'class':'woocommerce-loop-product__title'}).get_text()
price = cards.find('span', {'class':'price'}).get_text()
print(name)
print(price)
```

```python
#For all cases.
name = cards.find_all('h2', {'class':'woocommerce-loop-product__title'})
price = cards.find_all('span', {'class':'price'})
```

```python
names = []
prices = []
for i in name:
  names.append(i.get_text())

for j in price:
  prices.append(j.get_text())
```

```python
#Create pandas dataframe
data = pd.DataFrame({
    'name': names,
    'price': prices
    })
```

```python
#View resulting dataset
display(data)
```
