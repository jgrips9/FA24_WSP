```python
#Put into a function
def extract_data():
    # set the target URL
    chrome_options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(options=chrome_options)
    url = "https://scrapeme.live/shop/"
    driver.get(url)
    #Using xpath. Allows look for tag and attribute of any kind.
    pagecount = 1
    links = []
    name = []
    description = []
    price = []
    stock = []
    while pagecount < 5:
        link = driver.find_elements(By.XPATH, "//a[@class='woocommerce-LoopProduct-link woocommerce-loop-product__link']")
#Click on each link
        for item in range(len(link)):
            linkt = driver.find_elements(By.XPATH, "//a[@class='woocommerce-LoopProduct-link woocommerce-loop-product__link']")[item]
            linkt.click()
            name.append(driver.find_element(By.XPATH, "//h1[@class='product_title entry-title']").text)
            description.append(driver.find_element(By.XPATH, "//div[@class='woocommerce-product-details__short-description']").text)
            price.append(driver.find_element(By.XPATH, "//p[@class='price']").text)
            stock.append(driver.find_element(By.XPATH, "//p[@class='stock in-stock']").text)
            driver.back()
        pagecount = pagecount + 1
        nextp = driver.find_element(By.XPATH, "//a[@class='next page-numbers']")
        nextp.click()
    #Combine dataset
    data = pd.DataFrame({
    'name': name,
    'price': price,
    'description': description,
    'stock': stock
    })
    return data
```

```python
dataset = extract_data()
```
